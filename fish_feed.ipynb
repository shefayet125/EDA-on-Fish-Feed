```python
sns.set_style('whitegrid')
sns.pairplot(fish_feed)
```

```python
plt.figure(figsize=(15,7))
sns.barplot(x='Raw Materials',y='Before Covid',data=fish_feed,hue='Company',palette='Set1')
plt.legend(bbox_to_anchor=(1, 1), loc='upper left')
for i in plt.gca().patches:
 height, width = i.get_height(),i.get_x() + i.get_width()/2
 plt.annotate(f'{height}',(width, height),ha='center',va='bottom')
```

```python
plt.figure(figsize=(15,7))
sns.barplot(x='Raw Materials',y='During Covid',data=fish_feed,hue='Company',palette='Set1')
plt.legend(bbox_to_anchor=(1, 1), loc='upper left')
for i in plt.gca().patches:
 height, width = i.get_height(),i.get_x() + i.get_width()/2
 plt.annotate(f'{height}',(width, height),ha='center',va='bottom')
```

```python
plt.figure(figsize=(15,7))
sns.barplot(x='Raw Materials',y='After Covid',data=fish_feed,hue='Company',palette='Set1')
plt.legend(bbox_to_anchor=(1, 1), loc='upper left')
for i in plt.gca().patches:
 height, width = i.get_height(),i.get_x() + i.get_width()/2
 plt.annotate(f'{height}',(width, height),ha='center',va='bottom')
```

```python
pivot_df = fish_feed.pivot(index='Raw Materials', columns='Company')
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(pivot_df, cmap='coolwarm', annot=True, fmt=".1f", linewidths=.5)
plt.title('Raw Materials vs Company Heatmap')
plt.xlabel('Company')
plt.ylabel('Raw Materials')
plt.show()
```
